We import the dependencies into our Code

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

We Create a Pandas dataframe pointing to our iris data in the data folder,which will have no header so we give every column a name.

```python
df = pd.read_csv("data/iris.data", header=None, names=["sepal_length","sepal_width","petal_length","petal_width","species"])
```

Now this first separation of data is to tell the model that we gonna use: "Using this INPUTS you should give me this TARGET".                           

INPUTS = we take every column except "species" (axis=1 just means take in column form).                                                                       

TARGETS = we only take the "species" column.

```python
inputs = df.drop("species", axis=1)
targets = df["species"]
```

The next Code is another separation but this is different cause we need a data to train the model on (TRAIN DATA),and a data in which we will see how good the model can generalize (TEST DATA),the test size is self explanatory and the meaning of 0.2 is the percentage of data that we gonna use to test(20%),test_size=1 being the 100%.

Random_state=67 is for the shuffle of the data,like a seed,so it can be replicated.

It will give us 4 variables in the order that we see down here.

```python
train_x,test_x,train_y,test_y = train_test_split(inputs,targets,test_size=0.2,random_state=67)
```

Now we select the model that we gonna use,this model use the value of the 3 neighbors closest to the test_x that we gonna give to it later and make a prediction according to that.

For example if the closest 3 neighbors 2 of them are Versicolor and 1 is Setosa, the prediction of the model to the test gonna be Versicolor.

n_neighbors=3 is the amount of neighbors which the model gonna comparate the test data.

```python
model = KNeighborsClassifier(n_neighbors=3)
```

Simple Step,Train the model with our Training Data.

```python
model.fit(train_x,train_y)
```

We give to the model the Test Data and put the Predictions into our variable "Predictions"

```python
predictions = model.predict(test_x)
```

Now we give to the accuracy_score the corrects answers and the prediction of the model that we have trained and put them into the "Accuracy" variable.

```python
Accuracy = accuracy_score(test_y,predictions)
```

Print the Results

```python
print("Accuracy:",Accuracy)
```
